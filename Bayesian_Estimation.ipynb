```python
#Importing necessary libraries
from fractions import Fraction
Storage = [[[0 for j in range(0)] for i in range(0)]]
def counter(xAttributesValue, AttVal):
    count = 0
    for e in xAttributesValue:
        if e == AttVal:
            count = count + 1
    return count
```

```python
#Defining estimation function
def EstimateCalculator(
    xClassSet,
    xClassValue,
    xClassDistinctOccurence,
    xAttributesList,
    xAttributesSetListList,
    xAttributesValue,
    xAttValTranspose,
    xOccurenceLabel,
    xOccurenceValue,
    xnoOfTrainingData,
):
    num = xnoOfTrainingData * len(xClassSet)
    x = 0

    for i in range(len(xClassSet)):
        if len(Storage) > num:
            break

        Storage[x].append("P({})".format(list(xClassSet)[i]))
        n = str(xClassDistinctOccurence[i])
        d = str(len(xClassValue))
        Storage[x].append(Fraction(n + "/" + d))
        n = str(xClassDistinctOccurence[i] + 1)
        d = str(len(xClassValue) + len(xClassSet))
        Storage[x].append(Fraction(n + "/" + d))
        n = ""
        d = ""

        j = 0
        while j < len(xAttributesList):
            if len(Storage) > num:
                break
            for k in range(len(xAttributesSetListList)):
                if len(Storage) > num:
                    break
                x = x + 1
                Storage.append([])
                for m in range(len(xAttributesSetListList[k])):

                    Storage[x].append(
                        "P[({}={})|{}]".format(
                            xAttributesList[j],
                            xAttributesSetListList[k][m],
                            list(xClassSet)[i],
                        )
                    )

                    cnt = 0
                    for a in range(len(xClassValue)):
                        if (
                            xAttValTranspose[j][a] == xAttributesSetListList[k][m]
                            and xClassValue[a] == list(xClassSet)[i]
                        ):
                            cnt = cnt + 1

                    n = str(cnt)
                    d = str(xClassDistinctOccurence[i])
                    Storage[x].append(Fraction(n + "/" + d))
                    n = str(cnt + 1)
                    d = str(xClassDistinctOccurence[i] + len(xAttributesSetListList[k]))
                    Storage[x].append(Fraction(n + "/" + d))
                j = j + 1
```
